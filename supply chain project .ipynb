```python
# import pandas and numpy library 
import pandas as pd
import numpy as np
import math
```

```python
# order report
order_report = pd.read_csv("C:/Users/Mankirat/Downloads/Company X - Order Report.xlsx - Sheet1.csv")
# SKU master 
sku_master = pd.read_csv("C:/Users/Mankirat/Downloads/Company X - SKU Master.xlsx - Sheet1.csv")
# Pincodes Zones
pincode_zones = pd.read_csv("C:/Users/Mankirat/Downloads/Company X - Pincode Zones.xlsx - Sheet1.csv")
# Couries Invoice
courier_invoice = pd.read_csv("C:/Users/Mankirat/Downloads/Courier Company - Invoice.xlsx - Sheet1.csv")
# courier rates
courier_rates = pd.read_csv("C:/Users/Mankirat/Downloads/Courier Company - Rates.xlsx - Sheet1.csv")
```

```python
#marging order report with SKU Master 
weight=pd.merge(order_report,sku_master)
weight

```

```python
# checking null values 
weight.isnull().sum()
```

```python
#adding coloum
weight['total weight']=weight['Order Qty']*weight['Weight (g)']
#weight per order 
weight_per_order=weight.groupby('ExternOrderNo')['total weight'].sum().reset_index()
```

```python
#checking 
weight_per_order
```

```python
#Converting weight grams to kilograms 
weight_per_order['Total Weight_KG']=weight_per_order['total weight']/1000
```

```python
#Converting weight grams to kilograms 
weight_per_order
```

```python
#checking total weight for slab 
weight_per_order.max()
```

```python
# Function to determine the weight slab
def determine_weight_slab(weight):
    if weight <= 0.5:
        return 0.5
    else:
        return np.ceil(weight * 2) / 2
```

```python
# adding new column 
weight_per_order['Weight_slab_kg']=weight_per_order['Total Weight_KG'].apply(determine_weight_slab)
```

```python
weight_per_order
```

```python
#Merge with pincode zones to get the delivery zone 
order_zone=pd.merge(weight_per_order,courier_invoice,left_on='ExternOrderNo', right_on='Order ID', how='left')
order_zone=pd.merge(order_zone,pincode_zones,left_on=['Warehouse Pincode', 'Customer Pincode'], right_on=['Warehouse Pincode', 'Customer Pincode'], how='left')
```

```python
# run to check the code
order_zone
```

```python
# checking total number of duplicates 
order_zone.duplicated().sum()
```

```python
#droping duplicated and reseting index
order_zone.drop_duplicates(inplace=True)
order_zone.reset_index(drop=True,inplace=True)
```

```python
order_zone
```

```python
# merging order_zone with courier_rates 
order_rates = pd.merge(order_zone, courier_rates, how='cross')
```

```python
order_rates
```

```python
# Merge with Courier Rates to get the rates
order_with_rates = pd.merge(order_zone, courier_rates, how='cross')
# Calculate the expected charges
def calculate_expected_charges(row):
    if row['Type of Shipment'] == 'Forward charges':
        fixed_col = f"fwd_{row['Zone_y']}_fixed"
        additional_col = f"fwd_{row['Zone_y']}_additional"
    elif row['Type of Shipment'] == 'Forward and RTO charges':
        fixed_col = f"fwd_{row['Zone_y']}_fixed"
        additional_col = f"fwd_{row['Zone_y']}_additional"
        rto_fixed_col = f"rto_{row['Zone_y']}_fixed"
        rto_additional_col = f"rto_{row['Zone_y']}_additional"
    else:
        return 0
    
    # Calculate the number of additional slabs
    additional_slabs = (row['Weight_slab_kg'] - 0.5) / 0.5
    
    # Calculate the total charges
    if row['Type of Shipment'] == 'Forward charges':
        total_charges = row[fixed_col] + (additional_slabs * row[additional_col])
    else:
        total_charges = row[fixed_col] + (additional_slabs * row[additional_col]) + row[rto_fixed_col] + (additional_slabs * row[rto_additional_col])
    
    return total_charges

order_with_rates['Expected Charge (Rs.)'] = order_with_rates.apply(calculate_expected_charges, axis=1)
```

```python
order_with_rates
```

```python
# making a new column of Difference(Rs)
order_with_rates['Difference(Rs)']=order_with_rates['Expected Charge (Rs.)']-order_with_rates['Billing Amount (Rs.)']
```

```python
order_with_rates
```

```python
result= order_with_rates[['Order ID', 'AWB Code', 'Total Weight_KG', 'Weight_slab_kg', 'Charged Weight', 'Zone_x','Zone_y','Type of Shipment','Expected Charge (Rs.)', 'Billing Amount (Rs.)','Difference(Rs)']].round(2)

# Saving the result DataFrame to an Excel file
# result.to_excel('Result.xlsx', index=False)
```

```python
result
```

# Classified each order into 'Correct', 'Overcharged', or 'Undercharged' by comparing the actual billing amount with the expected charge."

```python
result['Billing_stutus'] = 'Correct'
result.loc[result['Difference(Rs)'] < 0, 'Billing_stutus'] = 'Overcharged'
result.loc[result['Difference(Rs)'] > 0,'Billing_stutus'] = 'Undercharged'
```

```python
result
```

# this summary table breaks down the billing accuracy for all orders. It classifies each order into one of three categories — 'Correct Charges', 'Overcharged', or 'Undercharged' — based on the difference between the actual billed amount and the expected charge.

```python
summary = pd.DataFrame({
    'Category': ['Correct Charges', 'Overcharged', 'Undercharged'],
    'Order Count': [
        (result['Billing_stutus'] == 'Correct').sum(),
        (result['Billing_stutus'] == 'Overcharged').sum(),
        (result['Billing_stutus'] == 'Undercharged').sum()
    ],
    'Amount (Rs.)': [
        result[result['Billing_stutus'] == 'Correct']['Billing Amount (Rs.)'].sum(),
        abs(result[result['Billing_stutus'] == 'Overcharged']['Difference(Rs)'].sum()),
        result[result['Billing_stutus'] == 'Undercharged']['Difference(Rs)'].sum()
    ]
})
```
